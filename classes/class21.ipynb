```python
r=range(1,5)
x=np.array(r)

x=np.linspace(0,2,100)
fig=plt.figure()
ax = fig.add_subplot(2,2,1)

```

```python
ax.plot(x)
ax.plot(x**2)
ax.plot(x**3)
fig   
```

```python
ax.plot(x, label = 'Linear')
ax.plot(x**2, label = 'Quadratic')
ax.plot(x**3, label = 'Cubic')
ax.set_xlabel('Year')
ax.set_ylabel('Amount')
ax.set_title('Sample plot')
ax.legend()

fig
```
